```python
domain = 'kinopoisk'
url = 'https://api.vk.com/method/wall.get'
params = {
          'access_token':'b614120ab614120ab614120adab5052917bb614b614120ad54ebb17a09cb7d575b0e0cb',
            'v':'5.131',
          'domain':domain,
          'count':1
            }
inf_0 = get (url, params = params) 
posts_count = inf_0.json() ['response'] ['count']
offsets_list = list(range (0,posts_count+1,100))
params ['count'] = 100

id_list = []
date_list = []
text_list = []
comment_list = []
like_list = []
repost_list = []
view_list = []

for offset in tqdm(offsets_list):
    params ['offset'] = offset
    inf_0 = get (url, params = params) 
    for post in inf_0.json() ['response'] ['items']:
        if 'id' and 'date' and 'text' and 'comments' and 'likes' and 'reposts' and 'views' in list(post.keys()):
            id_list.append (post ['id'])
            date_list.append (post ['date'])
            text_list.append (post ['text'])
            comment_list.append (post ['comments']['count'])
            like_list.append (post ['likes']['count'])
            repost_list.append (post ['reposts']['count'])
            view_list.append (post ['views']['count'])
```

```python
from nltk.tokenize import word_tokenize
```

```python
vk_tokens = []
for element in text_list:
    vk_tokens.extend (word_tokenize(element.lower()))
```

```python
vk_tokens
```

```python
import nltk
nltk.download ('averaged_perceptron_tagger_ru')
```

```python
from nltk import pos_tag
vk_tokens_tags = pos_tag(vk_tokens, lang = 'rus')
```

```python
vk_tokens_tags
```

```python
noun = ['S-PRO', 'S']
verb = ['V']
adj = ['A=comp2', 'A=m', 'A-PRO=f', 'A-PRO=pl', 'A-PRO', 'A=comp', 'A=sg', 'A', 'A=pl', 'A=f', 'A=n', 'A-PRO=n']
```

```python
from nltk.stem import WordNetLemmatizer

#vk_tokens_tags

vk_tokens_lem = []

for element in vk_tokens_tags:
    if element [1] in noun:
        vk_tokens_lem.append (WordNetLemmatizer().lemmatize (element [0], pos = 'n'))
    if element [1] in verb:
        vk_tokens_lem.append (WordNetLemmatizer().lemmatize (element [0], pos = 'v'))
    if element [1] in adj:
        vk_tokens_lem.append (WordNetLemmatizer().lemmatize (element [0], pos = 'a'))
```

```python
vk_tokens_lem
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words ('russian')
```

```python
stopwords
```

```python
tokens_final = []

for token in vk_tokens_lem:
    if token not in stopwords:
        tokens_final.append (token)
```

```python
tokens_final
```

```python
from collections import Counter
bag_of_words = Counter (tokens_final)
```

```python
print (bag_of_words)
```

```python
from nltk import SnowballStemmer

tokens_final_final = []

for token in tokens_final:
    tokens_final_final.append(SnowballStemmer ('russian').stem (token))
```

```python
tokens_final_final
```

```python
from collections import Counter
bag_of_words = Counter (tokens_final_final)
```
